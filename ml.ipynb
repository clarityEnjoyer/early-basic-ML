```python
m,n = X.shape
# w=np.random.rand(n)
w=np.zeros(n)
b=0
```

```python
w
```

```python
def J(X,y,w,b):
  cost=0
  #for i in range(m):
  #  cost += (np.dot(X[i],w)+b-y[i])**2
  cost=np.sum((np.dot(X,w)+b-y)**2)
  return cost/(2*m)
```

```python
J(X,y,w,b)
```

```python
def derivative(X,y,w,b):
  dj_dw=np.zeros(n)
  dj_db=0
  dj_db=(np.sum(np.dot(X,w)+b-y))/m
  for i in range(m):
    dj_dw += (np.dot(X[i],w)+b-y[i])*X[i]
  dj_dw/=m
  return dj_dw, dj_db
```

```python
print(w)
print(derivative(X,y,w,b))
```

```python
def step(X,y,w,b,alpha):
  dj_dw, dj_db=derivative(X,y,w,b)
  w=w-alpha*dj_dw
  b=b-alpha*dj_db
  return w,b
```

```python
step(X,y,w,b,1e-8)
```

```python
def gradient_descent(X, y, w, b, alpha, iterations, rate):
  cost_history=[]
  for it in range(iterations):
    cur = J(X,y,w,b)
    cost_history.append(cur)
    if (it%rate==0):
      print(f"{cur:e}")
      print(f"{w}") # Removed :e from here
    w,b = step(X,y,w,b,alpha)
  return w, b, cost_history
```

to by≈Ço bez skalowania danych

```python
def scaling1(X):
  maxi = np.max(X, axis=0)
  mini = np.min(X, axis=0)
  return (X-mini)/(maxi-mini)
```

```python
def scaling2(X):
  mean = np.mean(X, axis=0)
  std = np.std(X, axis=0)
  return (X-mean)/std
```

```python
def gradient_descent_scaled(X, y, w, b, alpha, iterations, rate,scalef):
  X_scaled=scalef(X)
  cost_history = []  # List to store cost at each iteration
  for it in range(iterations):
    cur = J(X_scaled,y,w,b)
    cost_history.append(cur)  # Append cost to history
    if (it%rate==0): # Added this if condition
      print(f"{cur:e}")
      print(f"{w}") # Removed :e from here
    w,b = step(X_scaled,y,w,b,alpha)
  return w, b, cost_history # Return cost history as well
```

```python
print("ze skalowaniem1")
print('-'*50)
w_final, b_final, cost_history = gradient_descent_scaled(X, y, w, b, 5e-2, 10000, 500,scaling1)

plt.plot(cost_history)
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.title("Cost vs. Iterations")
plt.show()
print('-'*50)

print("bez skalowania")
print('-'*50)
w_final, b_final, cost_history = gradient_descent(X, y, w, b, 1e-7, 10000, 500)

plt.plot(cost_history)
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.title("Cost vs. Iterations")
plt.show()
print('-'*50)

print("ze skalowaniem2")
print('-'*50)
w_final, b_final, cost_history = gradient_descent_scaled(X, y, w, b, 5e-2, 10000, 500,scaling2)

plt.plot(cost_history)
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.title("Cost vs. Iterations")
plt.show()
print('-'*50)
```
